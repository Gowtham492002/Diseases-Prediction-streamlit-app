```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
```

```python
df=pd.read_csv("/content/diabetes.csv")
```

```python
df.head()
```

```python
df.shape
```

```python
df.describe()
```

```python
df['Outcome'].value_counts()
```

```python
df.groupby('Outcome').mean()
```

```python
X=df.drop(columns="Outcome",axis=1)
y=df['Outcome']
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,stratify=y,random_state=1)
```

```python
classifier=svm.SVC(kernel="linear")
```

```python
classifier.fit(X_train,y_train)
```

```python
training_prediction=classifier.predict(X_train)
```

```python
training_accuracy=accuracy_score(training_prediction,y_train)
training_accuracy
```

```python
prediction=classifier.predict(X_test)
accuracy=accuracy_score(prediction,y_test)
accuracy
```

```python
input_data=(0,137,40,35,168,43.1,2.288,33)

inp_np=np.asarray(input_data)

inp_rshpd=inp_np.reshape(1,-1)

```

```python
prediction=classifier.predict(inp_rshpd)
if prediction[0]==0:
  print("The person is not diabetic")
else:
  print("The person is diabetic")
```
